```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Bài 2: Vẽ đồ thị đường đơn giản y = [1, 2, 3, 4] theo x = [1, 2, 3, 4].
# Bài 3: Thêm tiêu đề cho đồ thị trên.
# Bài 4: Gắn nhãn trục hoành là “X axis” và trục tung là “Y axis”.
# Bài 5: Thêm lưới (grid) vào đồ thị.
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
plt.title("Đồ thị hàm số y theo x")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python

# Bài 6: Vẽ hai đường trên cùng một đồ thị: y = x và y = x^2

x = [1, 2, 3, 4,  5, 6, 7]
y1 = x
y2 = [val**2 for val in x]

plt.plot(x, y1)
plt.plot(x, y2)
plt.title('Đồ thị hai hàm số')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bài 7: Đổi màu đường sang màu đỏ.
# Bài 8: Thay đổi kiểu đường (dotted, dashed).
# Bài 9: Thêm marker (dấu chấm tròn) cho đường.
# Bài 10: Lưu đồ thị ra file ảnh định dạng PNG

x = [1, 2, 3, 4]
y = [val**3 for val in x]
plt.title("Đồ thị hàm số y theo x")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y, color = 'red', linestyle ="--", marker = "o")
plt.grid(True)
plt.show()

plt.savefig("Đồ thị y = x^3")
```

```python
# Bài 11: Vẽ biểu đồ cột (bar chart) với dữ liệu số lượng sinh viên trong 4 lớp.
# Bài 12: Đặt màu sắc khác nhau cho từng cột.
# Bài 13: Thêm nhãn tên lớp dưới trục hoành.

x = ["Class A", "Class B", "Class C", "Class D"]

y = [23, 27, 15, 29]
plt.bar(x,y, color=("pink", "lightgreen", "yellow", "lightblue"))
plt.legend()
plt.show()
```

```python
# Bài 14: Vẽ biểu đồ cột ngang (horizontal bar chart).
x = ["Class A", "Class B", "Class C", "Class D"]
y = [23, 27, 15, 29]
plt.barh(x,y, color=("pink", "lightgreen", "yellow", "lightblue"))
plt.show()
```

```python
# Bài 15: Vẽ biểu đồ histogram với dữ liệu ngẫu nhiên chuẩn.
# Bài 16: Thay đổi số lượng bins trong histogram.
# Bài 17: Thêm đường dọc chỉ vị trí trung bình vào histogram.

data = np.random.randn(1000)
mean = np.mean(data)

plt.hist(data, bins = 100)
plt.axvline(mean, color = 'red')
plt.show()
```

```python
# Bài 18: Vẽ biểu đồ tán xạ (scatter plot) từ dữ liệu x và y ngẫu nhiên.
# Bài 19: Đặt màu sắc của điểm theo giá trị y (dùng cmap)

x = np.random.randn(1000)
y = np.random.uniform(-3, 3, 1000)
plt.scatter(x,y, c=y, cmap = "viridis", s = 20, alpha = 0.8)
plt.show()

from matplotlib import colors
norm = colors.Normalize(vmin=y.min(), vmax=y.max())
plt.scatter(x, y, c = y, cmap = 'plasma', norm = norm, s = 20)
plt.colorbar(label = 'y value')
```

```python
# Bài 20: Thay đổi kích thước điểm (marker size).
x = [1, 2, 3, 4]
y = [val**3 for val in x]
plt.title("Đồ thị hàm số y theo x")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y, color = 'red', linestyle ="--", marker = "o", markersize = 15)
plt.grid(True)
plt.show()

plt.savefig("Đồ thị y = x^3")
```

```python
# Bài 21: Thêm chú thích (legend) vào scatter plot
np.random.seed(1)
x = np.random.randn(100)
y = np.random.randn(100)
groups = np.random.choice(['A','B','C'], size=100)

plt.figure(figsize=(7,5))
for g in np.unique(groups):
    mask = groups==g
    plt.scatter(x[mask], y[mask], label=f"Group {g}", s=40, alpha=0.7)
plt.legend(title='Groups')
plt.show()
```

```python
# Bài 22: Vẽ biểu đồ hình tròn (pie chart) với tỉ lệ 4 phần
# Bài 23: Thêm nhãn cho từng phần của biểu đồ hình tròn
# Bài 24: Thêm hiệu ứng tách 1 phần (explode) trong pie chart
# Bài 25: Thay đổi màu sắc cho các phần của pie chart

sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
explode = (0, 0.2, 0, 0)
colors = ['red', 'yellow', 'green', 'purple']

plt.pie(sizes, labels = labels, explode = explode, colors = colors)
plt.show()
```

```python
# Bài 26: Vẽ nhiều subplot trong cùng một figure (2 hàng 2 cột).
# Bài 27: Vẽ đồ thị đường trong subplot (1)
# Bài 28: Vẽ histogram trong subplot (2).
# Bài 29: Vẽ scatter plot trong subplot (3).
# Bài 30: Vẽ pie chart trong subplot (4).

x = np.random.randn(100)
y1 = x
y3 = np.random.randn(100)

sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
explode = (0, 0.2, 0, 0)
colors = ['red', 'yellow', 'green', 'purple']

categories = ['A', 'B', 'C', 'D']
values = [23, 17, 35, 29]

fig, axs = plt.subplots(2, 2, figsize = (12, 8))

axs[0,0].plot(x, y1, color= 'tab:blue', label = 'y=x')
axs[0,1].hist(x, color = 'tab:orange')
axs[1,0].scatter(x, y3, color = 'tab:green', label= 'random data')

axs[1, 1].pie(sizes, labels = labels, colors = colors)
```

# Yêu cầu thêm: tự tìm data và vẽ ít nhất 4 biểu bồ, trong đó có 1 biểu đồ có subplot

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
print(df.head())
print(f"Shape: {df.shape}")
print(df.columns.tolist())
```

```python
np.random.seed(42)

survived_0 = df[df['survived']==0]
survived_1 = df[df['survived']==1]

fig, axs = plt.subplots(2, 2, figsize = (14, 10))

sex_survived = pd.crosstab(df['sex'], df['survived'])
sex_survived.plot(kind='bar', ax=axs[0,0], color=['red', 'green'], alpha=0.7)
axs[0,0].set_title('Sex Ratio by Survival')
axs[0,0].set_xlabel('Sex')
axs[0,0].set_ylabel('Count')
axs[0,0].set_xticklabels(['Female', 'Male'], rotation=0)
axs[0,0].legend(['Died', 'Survived'])

axs[0,1].hist(df['age'], color = 'tab:orange')
axs[0,1].set_title('Histogram of age')

axs[1,0].scatter(survived_0['age'], survived_0['fare'], color = 'tab:red', alpha = 0.6, label = 'Died', s= 50)
axs[1,0].set_title("Age vs Fare by Survival")
axs[1,0].scatter(survived_1['age'], survived_1['fare'], color = 'tab:blue', alpha = 0.6, label = 'Survived', s= 50)

class_counts = df['class'].value_counts()
axs[1,1].pie(class_counts, labels = class_counts.index, colors = ['tab:red', 'tab:green', 'tab:blue'], autopct='%1.1f%%')
axs[1,1].set_title("Class ratio", pad = 20)

plt.tight_layout()
plt.show()
```

```python
df.boxplot(column='age', by='survived', ax= plt.gca())
plt.title('Age distribution by Survival')
plt.suptitle('')
plt.xlabel('Survived (0=No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

```python
import matplotlib.patches as mpatches
fig, ax = plt.subplots(figsize=(8, 6))
numeric_df = df[['age', 'fare', 'survived', 'pclass']].corr()
im = ax.imshow(numeric_df, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
ax.set_xticks(range(len(numeric_df.columns)))
ax.set_yticks(range(len(numeric_df.columns)))
ax.set_xticklabels(numeric_df.columns, rotation=45)
ax.set_yticklabels(numeric_df.columns)
ax.set_title('Correlation Matrix')
plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
```
