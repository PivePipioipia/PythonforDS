```python
# Bài 1: Viết chương trình nhập vào một danh sách các số nguyên và in ra danh sách sau
# khi đã sắp xếp tăng dần.
number = input("Nhập vào danh sách số nguyên cách nhau dấu cách: ")
num_list = []
num_list = [int(x.strip()) for x in number.split(' ') if x.strip()]
num_list.sort()

print("Danh sách đã sắp xếp tăng dần: ", num_list)
```

```python
# Bài 2: Viết chương trình nhập vào một danh sách các số nguyên và in ra danh sách sau khi đã loại bỏ số trùng lặp.
numbers = [1, 5, 3, 2, 5, 1, 4, 3, 2, 8]

unique_numbers = list(set(numbers))

print(f"Danh sách ban đầu: {numbers}")
print(f"Danh sách sau khi loại bỏ trùng lặp: {unique_numbers}")
```

```python
from collections import Counter

# Bài 3: Viết chương trình nhập vào một danh sách và tìm giá trị xuất hiện nhiều nhất.
data = [4, 1, 5, 2, 4, 4, 3, 2, 1, 4]

counts = Counter(data)

most_common_item = counts.most_common(1)

if most_common_item:
    value, count = most_common_item[0]
    print(f"Danh sách: {data}")
    print(f"Giá trị xuất hiện nhiều nhất: {value} (Số lần: {count})")
else:
    print("Danh sách rỗng.")
```

```python
# Bài 4: Viết chương trình nhập vào một danh sách và in ra tất cả phần tử duy nhất (chỉ xuất hiện 1 lần).
data = [4, 1, 5, 2, 4, 4, 3, 2, 1, 4, 6]

counts = Counter(data)

unique_only = [item for item, count in counts.items() if count == 1]

print(f"Danh sách: {data}")
print(f"Các phần tử chỉ xuất hiện 1 lần: {unique_only}")
```

```python
# Bài 6: Viết chương trình đảo ngược một mảng mà không dùng hàm có sẵn (reverse() hoặc [::-1]).
data = [1, 2, 3, 4, 5]
reversed_data = []

for i in range(len(data) - 1, -1, -1):
    reversed_data.append(data[i])

print(f"Danh sách ban đầu: {data}")
print(f"Danh sách đảo ngược thủ công: {reversed_data}")
```

```python
# Bài 10: Viết chương trình nhập một danh sách và nhân đôi giá trị tất cả phần tử.
numbers = [1, 2, 3, 4, 5]

doubled_numbers = [x * 2 for x in numbers]

print(f"Danh sách ban đầu: {numbers}")
print(f"Danh sách sau khi nhân đôi: {doubled_numbers}")
```

```python
# Bài 11: Viết chương trình nhập một danh sách và thay thế tất cả số 0 bằng số 1.
numbers = [1, 0, 5, 0, 3, 0, 2]

replaced_numbers = [1 if x == 0 else x for x in numbers]

print(f"Danh sách ban đầu: {numbers}")
print(f"Danh sách sau khi thay thế 0 bằng 1: {replaced_numbers}")
```

```python
import math


# Bài 12: Viết chương trình nhập một danh sách và tính tích các phần tử.
numbers = [1, 2, 3, 4, 5]

product = math.prod(numbers)

print(f"Danh sách: {numbers}")
print(f"Tích các phần tử: {product}")
```

```python
# Bài 13: Viết chương trình nhập một danh sách và kiểm tra xem nó có đối xứng (palindrome) không.
list1 = [1, 2, 3, 2, 1]
list2 = [1, 2, 3, 4, 5]

is_palindrome1 = list1 == list1[::-1]
is_palindrome2 = list2 == list2[::-1]

print(f"List {list1} là Palindrome: {is_palindrome1}")
print(f"List {list2} là Palindrome: {is_palindrome2}")
```

```python
# Bài 14: Viết chương trình nhập một danh sách và tìm 2 phần tử có tổng lớn nhất.
numbers = [10, 5, 20, 8, 15]

numbers.sort(reverse=True)

max1 = numbers[0]
max2 = numbers[1]

print(f"Danh sách đã sắp xếp: {numbers}")
print(f"Hai phần tử có tổng lớn nhất: {max1} và {max2} (Tổng: {max1 + max2})")
```

```python
# Bài 15: Viết chương trình nhập một danh sách và tìm 2 phần tử có tổng nhỏ nhất.
numbers = [10, 5, 20, 8, 15]
numbers.sort()

min1 = numbers[0]
min2 = numbers[1]

print(f"Danh sách đã sắp xếp: {numbers}")
print(f"Hai phần tử có tổng nhỏ nhất: {min1} và {min2} (Tổng: {min1 + min2})")
```

```python
# Bài 16: Viết chương trình nhập vào một danh sách chuỗi và sắp xếp theo độ dài chuỗi.
strings = ["data", "science", "AI", "engineer", "python"]

sorted_by_length = sorted(strings, key=len)

print(f"Danh sách ban đầu: {strings}")
print(f"Danh sách sắp xếp theo độ dài: {sorted_by_length}")
```

```python
# Bài 17: Viết chương trình nhập vào một danh sách chuỗi và lọc ra các chuỗi bắt đầu bằng chữ cái ``a''.
strings = ["apple", "Banana", "cat", "algorithm", "Array", "zebra"]

filtered_strings = [
    s for s in strings
    if s.lower().startswith('a')
]

print(f"Danh sách ban đầu: {strings}")
print(f"Chuỗi bắt đầu bằng 'a' (không phân biệt): {filtered_strings}")
```

```python
# Bài 18: Viết chương trình nhập vào một danh sách số nguyên và xoá tất cả phần tử trùng lặp nhưng giữ nguyên thứ tự.
data = [1, 5, 3, 2, 5, 1, 4, 3, 2, 8]
result = []
seen = set()

for item in data:
    if item not in seen:
        seen.add(item)
        result.append(item)

print(f"Danh sách ban đầu: {data}")
print(f"Xoá trùng lặp, giữ nguyên thứ tự: {result}")
```

```python
# Bài 19: Viết chương trình nhập vào một danh sách số nguyên và in ra danh sách chỉ chứa các số nguyên tố.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

numbers = [2, 7, 10, 11, 15, 17, 20]

prime_numbers = [x for x in numbers if is_prime(x)]

print(f"Danh sách ban đầu: {numbers}")
print(f"Danh sách số nguyên tố: {prime_numbers}")
```

```python
# Bài 20: Viết chương trình nhập vào một danh sách số nguyên và in ra phần tử lớn thứ hai.
numbers = [10, 5, 20, 8, 15, 20]

unique_numbers = list(set(numbers))

unique_numbers.sort()

if len(unique_numbers) < 2:
    print("Không đủ phần tử để tìm phần tử lớn thứ hai.")
else:
    second_largest = unique_numbers[-2]
    print(f"Danh sách ban đầu: {numbers}")
    print(f"Danh sách không trùng lặp: {unique_numbers}")
    print(f"Phần tử lớn thứ hai: {second_largest}")
```
